```python
LAMBDA_1A = 0
LAMBDA_2A = 1
LAMBDA_3A = 2
EIGENV_1A_X = 3 
EIGENV_1A_Y = 4
EIGENV_1A_Z = 5
EIGENV_2A_X = 6 
EIGENV_2A_Y = 7
EIGENV_2A_Z = 8
EIGENV_3A_X = 9 
EIGENV_3A_Y = 10
EIGENV_3A_Z = 11
LAMBDA_1B = 12
LAMBDA_2B = 13
LAMBDA_3B = 14
EIGENV_1B_X = 15
EIGENV_1B_Y = 16
EIGENV_1B_Z = 17
EIGENV_2B_X = 18
EIGENV_2B_Y = 19
EIGENV_2B_Z = 20
EIGENV_3B_X = 21 
EIGENV_3B_Y = 22
EIGENV_3B_Z = 23

HALO_XA = 0
HALO_YA = 1
HALO_ZA = 2
HALO_MASS_A=3
HALO_XB = 4
HALO_YB = 5
HALO_ZB = 6
HALO_MASS_B=7
HALO_NORMAL_X=8
HALO_NORMAL_Y=9
HALO_NORMAL_Z=10
HALO_ECC=14

PEAK = 3
FILAMENT = 2
SHEET = 1
VOID = 0
environments = ["void", "sheet", "filament", "peak"]
```

```python
def distance(xA, yA, zA, xB, yB, zB):
    dist = (xA-xB)**2 + (yA-yB)**2 + (zA-zB)**2
    dist = sqrt(dist)
    return dist
```

```python
def mu_angle(xA, yA, zA, xB, yB, zB):
    mu = xA*xB + yA*yB + zA*zB
    normA = sqrt(xA**2 + yA**2 + zA**2)
    normB = sqrt(xB**2 + yB**2 + zB**2)
    mu = mu/(normA * normB)
    return mu
```

```python
def ellipticity(lambda1, lambda2, lambda3):
    DELTA = lambda1 + lambda2 + lambda3
    ELL = (lambda1 - lambda3)/(2.0*DELTA)
    return ELL
```

```python
def prolateness(lambda1, lambda2, lambda3):
    DELTA = lambda1 + lambda2 + lambda3
    PROL =  (lambda1 + lambda3 - 2.0*lambda2)/(2.0*DELTA)
    return PROL
```

```python
def environment_type(lambda1, lambda2, lambda3, threshold=0.3):
    n_points =  size(lambda1)
    env_type = int_(zeros(n_points)) * -1
    
    index = where((lambda3>threshold))
    env_type[index] = PEAK
    
    index = where((lambda2>threshold)&(lambda3<threshold))
    env_type[index] = FILAMENT
    
    index = where((lambda1>threshold)&(lambda2<threshold))
    env_type[index] = SHEET
    
    index = where(lambda1<threshold)
    env_type[index] = VOID
    
    return env_type
```

```python
web_data_2S = loadtxt("../data/2sigma/Tweb256_pairs2s.dat")
halo_data_2S = loadtxt("../data/2sigma/data_pairs2s.txt")

web_data_3S = loadtxt("../data/3sigma/Tweb256_pairs3s.dat")
halo_data_3S = loadtxt("../data/3sigma/data_pairs3s.txt")

web_data_general = loadtxt("../data/general/Tweb256_lgsample_vrn.dat")
halo_data_general = loadtxt("../data/general/data_lgsample_vrn.txt")
```

```python
def median_in_mass_bin(halo_data_ALL, web_data_ALL, min_mass=0.0, max_mass=100.0):
    total_mass_ALL = log10(halo_data_ALL[:,HALO_MASS_B] + halo_data_ALL[:,HALO_MASS_A])
    index = where((total_mass_ALL>min_mass)&(total_mass_ALL<max_mass))
    index = index[0]
  
    total_mass = log10(halo_data_ALL[index,HALO_MASS_B] + halo_data_ALL[index,HALO_MASS_A])
    ecc_ALL = halo_data_ALL[index,HALO_ECC]
    
    norm_delta_ALL = web_data_ALL[index,LAMBDA_1A] + web_data_ALL[index, LAMBDA_2A] + web_data_ALL[index, LAMBDA_3A]

    ELL_halo_A_ALL = ellipticity(web_data_ALL[index,LAMBDA_1A], web_data_ALL[index, LAMBDA_2A], web_data_ALL[index, LAMBDA_3A]) 

    PROL_halo_A_ALL = prolateness(web_data_ALL[index,LAMBDA_1A], web_data_ALL[index, LAMBDA_2A], web_data_ALL[index, LAMBDA_3A]) 

    env_halo_A_ALL = environment_type(web_data_ALL[index,LAMBDA_1A], web_data_ALL[index, LAMBDA_2A], web_data_ALL[index, LAMBDA_3A])

    mu_norm_eigenv1_ALL = mu_angle(halo_data_ALL[index,HALO_NORMAL_X], halo_data_ALL[index,HALO_NORMAL_Y], halo_data_ALL[index,HALO_NORMAL_Z], 
                          web_data_ALL[index, EIGENV_1A_X], web_data_ALL[index, EIGENV_1A_Y], web_data_ALL[index, EIGENV_1A_Z])

    mu_norm_eigenv2_ALL = mu_angle(halo_data_ALL[index,HALO_NORMAL_X], halo_data_ALL[index,HALO_NORMAL_Y], halo_data_ALL[index,HALO_NORMAL_Z], 
                          web_data_ALL[index, EIGENV_2A_X], web_data_ALL[index, EIGENV_2A_Y], web_data_ALL[index, EIGENV_2A_Z])

    mu_norm_eigenv3_ALL = mu_angle(halo_data_ALL[index,HALO_NORMAL_X], halo_data_ALL[index,HALO_NORMAL_Y], halo_data_ALL[index,HALO_NORMAL_Z],
                          web_data_ALL[index, EIGENV_3A_X], web_data_ALL[index, EIGENV_3A_Y], web_data_ALL[index, EIGENV_3A_Z])
    med_0 = mean(total_mass)
    med_1 = median(abs(mu_norm_eigenv1_ALL))
    med_2 = median(abs(mu_norm_eigenv2_ALL))
    med_3 = median(abs(mu_norm_eigenv3_ALL))
    med_4 = mean(norm_delta_ALL)
    med_5 = mean(ELL_halo_A_ALL)
    med_6 = mean(abs(PROL_halo_A_ALL))
    med_7 = mean(ecc_ALL)
    sigma_0 = std(total_mass)
    sigma_4 = std(norm_delta_ALL)
    sigma_5 = std(ELL_halo_A_ALL)
    sigma_6 = std(abs(PROL_halo_A_ALL))
    sigma_7 = std(ecc_ALL)
    
    n_peak = size(where(env_halo_A_ALL==PEAK))
    n_filament = size(where(env_halo_A_ALL==FILAMENT))
    n_sheet = size(where(env_halo_A_ALL==SHEET))
    n_void = size(where(env_halo_A_ALL==VOID))
    n_total = 1.0*size(env_halo_A_ALL)
    print "min_mass, max_mass", min_mass, max_mass
    print n_peak, n_filament, n_sheet, n_void
    print n_peak/n_total, n_filament/n_total, n_sheet/n_total, n_void/n_total

    return med_0, sigma_0, med_1, med_2, med_3, med_4, sigma_4, med_5, sigma_5, med_6, sigma_6, med_7, sigma_7, n_peak, n_filament, n_sheet, n_void, n_total
```
