```python
def n_halos(total_mass, min_mass=1.0, max_mass=100.):
    index = where((total_mass<max_mass)&(total_mass>min_mass))
    n_tot = size(index)
    return n_tot
```

```python
n_bins = 24
log_mass_bins = linspace(11.5,13.1,n_bins+1)
print log_mass_bins
log_mass_center = 0.5*(log_mass_bins[1:] + log_mass_bins[:-1])
```

```python
tot_mass_distro_ALL = ones(n_bins)
tot_mass_distro_3S = ones(n_bins)
tot_mass_distro_NEW = ones(n_bins)
for i in arange(n_bins):
    tot_mass_distro_ALL[i] = n_halos(total_mass_ALL, min_mass=log_mass_bins[i], max_mass=log_mass_bins[i+1])
    tot_mass_distro_3S[i] = n_halos(total_mass_3S, min_mass=log_mass_bins[i], max_mass=log_mass_bins[i+1])
```

```python
plot(log_mass_center, tot_mass_distro_ALL)
plot(log_mass_center, tot_mass_distro_3S*5)
```

```python
def select_random_points(total_mass, min_mass=1.0, max_mass=2.0, nselect=1):
    index = where((total_mass<max_mass)&(total_mass>min_mass))
    index = index[0]
    random.shuffle(index)
    selected = index[0:nselect]
    return selected
```

```python
new_sample_ID = empty((0))
for i in arange(n_bins):
    n_in_3S  = n_halos(total_mass_3S, min_mass=log_mass_bins[i], max_mass=log_mass_bins[i+1])
    index = select_random_points(total_mass_ALL, min_mass=log_mass_bins[i], max_mass=log_mass_bins[i+1], nselect=n_in_3S*5)
    if(n_in_3S):
        new_sample_ID = append(new_sample_ID, index)
new_sample_ID = int_(new_sample_ID)
total_mass_NEW = total_mass_ALL[new_sample_ID]
```

```python
tot_mass_distro_ALL = ones(n_bins)
tot_mass_distro_3S = ones(n_bins)
tot_mass_distro_NEW = ones(n_bins)
for i in arange(n_bins):
    tot_mass_distro_ALL[i] = n_halos(total_mass_ALL, min_mass=log_mass_bins[i], max_mass=log_mass_bins[i+1])
    tot_mass_distro_3S[i] = n_halos(total_mass_3S, min_mass=log_mass_bins[i], max_mass=log_mass_bins[i+1])
    tot_mass_distro_NEW[i] = n_halos(total_mass_NEW, min_mass=log_mass_bins[i], max_mass=log_mass_bins[i+1])
```
