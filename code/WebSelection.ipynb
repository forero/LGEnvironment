```python
def distance(xA, yA, zA, xB, yB, zB):
    dist = (xA-xB)**2 + (yA-yB)**2 + (zA-zB)**2
    dist = sqrt(dist)
    return dist
```

```python
def mu_angle(xA, yA, zA, xB, yB, zB):
    mu = xA*xB + yA*yB + zA*zB
    normA = sqrt(xA**2 + yA**2 + zA**2)
    normB = sqrt(xB**2 + yB**2 + zB**2)
    mu = mu/(normA * normB)
    return mu
```

```python
def ellipticity(lambda1, lambda2, lambda3):
    DELTA = lambda1 + lambda2 + lambda3
    ELL = (lambda1 - lambda3)/(2.0*DELTA)
    return ELL
```

```python
def prolateness(lambda1, lambda2, lambda3):
    DELTA = lambda1 + lambda2 + lambda3
    PROL =  (lambda1 + lambda3 - 2.0*lambda2)/(2.0*DELTA)
    return PROL
```

```python
def environment_type(lambda1, lambda2, lambda3, threshold=0.3):
    n_points =  size(lambda1)
    env_type = int_(zeros(n_points)) * -1
    
    index = where((lambda3>threshold))
    env_type[index] = PEAK
    
    index = where((lambda2>threshold)&(lambda3<threshold))
    env_type[index] = FILAMENT
    
    index = where((lambda1>threshold)&(lambda2<threshold))
    env_type[index] = SHEET
    
    index = where(lambda1<threshold)
    env_type[index] = VOID
    
    return env_type
```

```python
def plot_alignment_e3(ENV=0, kind="r"):

    fig = figure(figsize=(9.5,6.5))
    ax = axes()
    label1="$\mathrm{2\sigma\  sample}$"
    label2="$\mathrm{3\sigma\  sample}$"
    label3="$\mathrm{Full\ sample}$"
    label_x="$\mu\equiv|\hat{e}_3\cdot\ \hat{%s}|$"%(kind)

    ax.set_xlabel(label_x,fontsize=25)
    ax.set_ylabel("${\mathrm{Fraction(<\mu)}}$",fontsize=25)

    enlarge_labels(ax)

    histo_bins = linspace(0,1.0,20)
    delta = 0.5*(histo_bins[1] - histo_bins[0])


    if(ENV>=0):
        env_id_ALL = where(env_halo_A_ALL==ENV)
        env_id_3S = where(env_halo_A_3S==ENV)
        env_id_2S = where(env_halo_A_2S==ENV)
    else:
        env_id_ALL = where(env_halo_A_ALL>ENV)
        env_id_3S = where(env_halo_A_3S>ENV)
        env_id_2S = where(env_halo_A_2S>ENV)
    ax.plot(sort(abs(mu_pos_eigenv3_ALL[env_id_ALL])), linspace(0,1.0,size(env_id_ALL)),color='black',ms=1,linewidth=3.0, label="$\mathrm{General\ Sample}$")
    ax.plot(sort(abs(mu_pos_eigenv3_3S[env_id_3S])), linspace(0,1.0,size(env_id_3S)),'r--',color='black',ms=1,linewidth=3.0, label="$3\sigma\ \mathrm{sample}$")
    ax.plot(sort(abs(mu_pos_eigenv3_2S[env_id_2S])), linspace(0,1.0,size(env_id_2S)),'r:',color='black',ms=1,linewidth=3.0, label="$2\sigma\ \mathrm{sample}$")

    line_x = linspace(0.0,1.0,100)
    ax.plot(line_x, line_x, color='black')

    ax.legend(loc=0, scatterpoints=1, prop={'size':20})

    ax.set_xlim([0.0,1.0])
    ax.set_ylim([0.0,1.0])
    if(ENV>=0):
        texto = "${\mathrm{%s}}$"%(environments[ENV])
        ax.set_title(texto, fontsize=35)
        filename = "alignments_subs_e3_"+kind+"_"+environments[ENV]
    else:
        texto = "${\mathrm{%s}}$"%("all\ environments")
        ax.set_title(texto, fontsize=35)
        filename = "alignments_subs_e3_"+kind+"_all_environments"
    fig.savefig(filename + '.eps',format = 'eps', transparent=True)
    fig.savefig(filename + '.pdf',format = 'pdf', transparent=True)
    fig.savefig(filename + '.png',format = 'png', transparent=True)
```
