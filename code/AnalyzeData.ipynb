```python
def distance(xA, yA, zA, xB, yB, zB):
    dist = (xA-xB)**2 + (yA-yB)**2 + (zA-zB)**2
    dist = sqrt(dist)
    return dist
```

```python
def mu_angle(xA, yA, zA, xB, yB, zB):
    mu = xA*xB + yA*yB + zA*zB
    normA = sqrt(xA**2 + yA**2 + zA**2)
    normB = sqrt(xB**2 + yB**2 + zB**2)
    mu = mu/(normA * normB)
    return mu
```

```python
def ellipticity(lambda1, lambda2, lambda3):
    DELTA = lambda1 + lambda2 + lambda3
    ELL = (lambda1 - lambda3)/(2.0*DELTA)
    return ELL
```

```python
def prolateness(lambda1, lambda2, lambda3):
    DELTA = lambda1 + lambda2 + lambda3
    PROL =  (lambda1 + lambda3 - 2.0*lambda2)/(2.0*DELTA)
    return PROL
```

```python
def environment_type(lambda1, lambda2, lambda3, threshold=0.3):
    n_points =  size(lambda1)
    env_type = int_(zeros(n_points)) * -1
    
    index = where((lambda3>threshold))
    env_type[index] = PEAK
    
    index = where((lambda2>threshold)&(lambda3<threshold))
    env_type[index] = FILAMENT
    
    index = where((lambda1>threshold)&(lambda2<threshold))
    env_type[index] = SHEET
    
    index = where(lambda1<threshold)
    env_type[index] = VOID
    
    return env_type
```

```python
def get_randoms(n_points=10):
    n_random = 5000
    n_points_x = 80
    y_axis = linspace(0.0,1.0,n_points)
    x_axis_dense = linspace(0.0,1.0,n_points_x)
    y_axis_dense_all = zeros([n_random,n_points_x])
    for i in range(n_random):
        ran_range = random.random(n_points)
        ran_range = sort(ran_range)
        ran_range[0] = 0.0
        ran_range[-1] = 1.0
        f = interp1d(ran_range, y_axis)
        y_axis_dense = f(x_axis_dense)
        y_axis_dense_all[i,:] = y_axis_dense[:]

    y_axis_median = ones(n_points_x)
    y_axis_low = ones(n_points_x)
    y_axis_high = ones(n_points_x)
    for i in range(n_points_x):
        y_axis_median[i] = median(y_axis_dense_all[:,i])
        y_axis_low[i] = percentile(y_axis_dense_all[:,i],5)
        y_axis_high[i] = percentile(y_axis_dense_all[:,i],95)

    return x_axis_dense, y_axis_median, y_axis_low, y_axis_high
```

```python
def enlarge_labels(ax, size=20):
    ticklabels_x = ax.get_xticklabels()
    ticklabels_y = ax.get_yticklabels()
    for label_x in ticklabels_x:
        label_x.set_fontsize(size)
        label_x.set_family('serif')
    for label_y in ticklabels_y:
        label_y.set_fontsize(size)
        label_y.set_family('serif')
```

```python
def plot_alignment_e3(ENV=0, kind="r"):

    fig = figure(figsize=(9.5,8.5))
    ax = axes()
    label1="$\mathrm{2\sigma\  sample}$"
    label2="$\mathrm{3\sigma\  sample}$"
    label3="$\mathrm{Full\ sample}$"
    label_x="$\mu\equiv|\hat{e}_3\cdot\ \hat{%s}|$"%(kind)

    ax.set_xlabel(label_x,fontsize=30)
    ax.set_ylabel("${\mathrm{Fraction(<\mu)}}$",fontsize=30)

    enlarge_labels(ax, size=30)

    histo_bins = linspace(0,1.0,20)
    delta = 0.5*(histo_bins[1] - histo_bins[0])

    x_random, y_random, y_random_up, y_random_down = get_randoms(n_points=n_points_2S)
    ax.fill_between(x_random, y_random_down, y_random_up, color='silver', alpha=0.6)
    ax.plot(x_random, y_random, color='black', linewidth=0.2)
    ax.plot(x_random, y_random_down, color='black', linewidth=0.2)
    ax.plot(x_random, y_random_up, color='black', linewidth=0.2)


    if(ENV>=0):
        env_id_ALL = where(env_halo_A_ALL==ENV)
        env_id_3S = where(env_halo_A_3S==ENV)
        env_id_2S = where(env_halo_A_2S==ENV)
    else:
        env_id_ALL = where(env_halo_A_ALL>ENV)
        env_id_3S = where(env_halo_A_3S>ENV)
        env_id_2S = where(env_halo_A_2S>ENV)
    ax.plot(sort(abs(mu_pos_eigenv3_ALL[env_id_ALL])), linspace(0,1.0,size(env_id_ALL)),color='black',ms=1,linewidth=3.0, label="$\mathrm{General\ Sample}$")
    ax.plot(sort(abs(mu_pos_eigenv3_3S[env_id_3S])), linspace(0,1.0,size(env_id_3S)),'r--',color='black',ms=1,linewidth=3.0, label="$3\sigma\ \mathrm{sample}$")
    ax.plot(sort(abs(mu_pos_eigenv3_2S[env_id_2S])), linspace(0,1.0,size(env_id_2S)),'r:',color='black',ms=1,linewidth=3.0, label="$2\sigma\ \mathrm{sample}$")

    ax.legend(loc=0, scatterpoints=1, prop={'size':30})

    ax.set_xlim([0.0,1.0])
    ax.set_ylim([0.0,1.0])
    if(ENV>=0):
        texto = "${\mathrm{%s}}$"%(environments[ENV])
        ax.set_title(texto, fontsize=35)
        filename = "alignments_e3_"+kind+"_"+environments[ENV]
    else:
        filename = "alignments_e3_"+kind+"_all_environments"
    fig.savefig(filename + '.eps',format = 'eps', transparent=True)
    fig.savefig(filename + '.pdf',format = 'pdf', transparent=True)
    fig.savefig(filename + '.png',format = 'png', transparent=True)
```
