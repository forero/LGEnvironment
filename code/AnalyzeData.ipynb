```python
def distance(xA, yA, zA, xB, yB, zB):
    dist = (xA-xB)**2 + (yA-yB)**2 + (zA-zB)**2
    dist = sqrt(dist)
    return dist
```

```python
def mu_angle(xA, yA, zA, xB, yB, zB):
    mu = xA*xB + yA*yB + zA*zB
    normA = sqrt(xA**2 + yA**2 + zA**2)
    normB = sqrt(xB**2 + yB**2 + zB**2)
    mu = mu/(normA * normB)
    return mu
```

```python
def fractional_anisotropy(lambda1, lambda2, lambda3):
    d13 = (lambda1 - lambda3)**2
    d23 = (lambda2 - lambda3)**2
    d12 = (lambda1 - lambda2)**2
    norm = lambda1**2 + lambda2**2 + lambda3**2
    FA = sqrt((d13 + d23 + d12)/norm)
    FA = 1.0/sqrt(3.0) * FA
    return FA
```

```python
def environment_type(lambda1, lambda2, lambda3, threshold=0.3):
    n_points =  size(lambda1)
    env_type = int_(zeros(n_points)) * -1
    
    index = where((lambda3>threshold))
    env_type[index] = PEAK
    
    index = where((lambda2>threshold)&(lambda3<threshold))
    env_type[index] = FILAMENT
    
    index = where((lambda1>threshold)&(lambda2<threshold))
    env_type[index] = SHEET
    
    index = where(lambda1<threshold)
    env_type[index] = VOID
    
    return env_type
```

```python
def plot_alignment(mu_data, id_data, 
                   label1="$\mathrm{Full\ 2\sigma\  sample}$", label2="$\mathrm{Pairs\ in\ sheets}$",
                   fileout="alignment_sheet_pos_2S", label_x="$|\hat{e}_3\cdot\ \hat{r}|$"):

    fig = figure(figsize=(9.5,6.5))
    ax = axes()
    ax.set_xlabel(label_x,fontsize=25)
    ax.set_ylabel("${\mathrm{Normalized\ histogram}}$",fontsize=25)
    ticklabels_x = ax.get_xticklabels()
    ticklabels_y = ax.get_yticklabels()

    for label_x in ticklabels_x:
        label_x.set_fontsize(20)
        label_x.set_family('serif')
    for label_y in ticklabels_y:
        label_y.set_fontsize(20)
        label_y.set_family('serif')

    histo_bins = linspace(0,1.0,9)
    delta = 0.5*(histo_bins[1] - histo_bins[0])

    hist_A, bin_edges_A = histogram(abs(mu_data), bins=histo_bins, density=True)
    hist_B, bin_edges_B = histogram(abs(mu_data[id_data]), bins=histo_bins, density=True)

    ax.plot(bin_edges_A[:-1]+delta,hist_A,color='black',ms=1,linewidth=3.0, label=label1)
    ax.plot(bin_edges_B[:-1]+delta,hist_B,'r--',color='black',ms=2,linewidth=3.0, label=label2)


    ax.legend(loc=0, scatterpoints=1, prop={'size':20})

    ax.set_xlim([0.0,1.0])
    ax.set_ylim([0.0,3.5])

    filename = fileout
    fig.savefig(filename + '.eps',format = 'eps', transparent=True)
    fig.savefig(filename + '.pdf',format = 'pdf', transparent=True)
    fig.savefig(filename + '.png',format = 'png', transparent=True)
```
